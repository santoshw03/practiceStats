```python
#with the help of center limit theorem we can convert a non-normal distibution data into a normal distribution data
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
population = [np.random.randint(10,100) for i in range(10000)] #to generate random 10000 numbers which lie between 1 to 100
len(population)
```

```python
pop_table = pd.DataFrame({"pop_data":population}) # to arrange the data into a table
pop_table
```

```python
sns.kdeplot(x="pop_data",data=pop_table)
plt.show()
```

```python
# from the graph we can see it is not a normal distribution data
```

```python
# so we first select random samples from population(size of sample should be greater than 30 and less than 10% of population)
# after getting samples we find mean of all samples and store them all together
# then we find the mean of all the means
```

```python
sample_means = []
for samples in range(50): # to get 50 samples
    sample_data = []
    for data in range(500): #as 500 is greater than 30 and less than 10% of population
        sample_data.append(np.random.choice(population)) #to add random numbers from popultion into the sample_data
    sample_means.append(np.mean(sample_data)) #to add the means of sample_data in sample_means 
        
```

```python
sample_m_table = pd.DataFrame({"sample_mean":sample_means}) #making table of smaple_means
```

```python
sns.kdeplot(x="sample_mean",data=sample_m_table)
plt.show()
```

```python
#it is a normal distribution data, we can confirm it by finding mean and median
np.mean(sample_means) , np.median(sample_means)
```
