```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# mean absolute deviation is used to know how much the data is spread from its mean 
# less the data is spread more important it is
# MAD = (sumof(|xi - meanof(xi)|))/n
```

```python
a = np.array([75,63,73,68,72,67])
b = np.array([90,47,43,96,93,51])
```

```python
no = np.array([1,2,3,4,5,6]) #created array for the y-axis
```

```python
mean_a = np.mean(a)
mean_b = np.mean(b)
mean_a,mean_b
```

```python
#we can see that the mean of both data are approx same but the spread of data varies ,, hence to find of which data is more reliable we use MAD when mean is approx same
```

```python
plt.figure(figsize=(10,3)) # to stretch the graph
plt.scatter(a,no)
plt.show()
```

```python
plt.figure(figsize=(10,3))
plt.scatter(b,no,color="red")
plt.show()
```

```python
#comparing
plt.figure(figsize=(10,3))
plt.scatter(a,no,label="A")
plt.scatter(b,no,color="red",label="B")
plt.legend()
plt.show()
```

```python
#comparing with mean
plt.figure(figsize=(10,3))
plt.scatter(a,no,label="A")
plt.scatter(b,no,color="red",label="B")
plt.plot([mean_b for i in range(0,7)],[i for i in range(0,7)])
plt.legend()
plt.show()
```

```python
#from the graph we can see that values of A are more closer to the mean than those of B
#which means A has less spread
#hence A is more reliable
```

```python
# lets find meand absolute deviiation MAD by using maths from the formula mentioned above
 
```

```python
#lets first find |xi - xi(bar)| means |xi - meanof(xi)|
diff_a = np.abs(a-a.mean())
diff_b = np.abs(b-b.mean())
```

```python
diff_a
```

```python
diff_b
```

```python
#now we can find mad by
mad_a = (np.sum(diff_a)/len(a))
mad_b = (np.sum(diff_b)/len(b))
```

```python
mad_a,mad_b
```

```python
#as mad_a < mad_b we can say that a is less spread and hence more reliable
```
